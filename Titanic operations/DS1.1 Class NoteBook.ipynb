```python
import csv 
with open('titanic.csv') as f:
    csv_reader = csv.reader(f, delimiter=',')
    count = 0 
    for row in csv_reader:
        if row[1] == '1' and row[4] == 'female':
            count += 1
    print('{} Women survived'.format(count))
```
