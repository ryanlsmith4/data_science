```python
cities = df.City.dropna()
city_dict = {}
for city in cities:
    if city in city_dict:
        city_dict[city] += 1
    else:
        city_dict[city] = 1
city_dict        
```

```python
maximum_location = max(city_dict, key=city_dict.get)
(maximum_location, city_dict[maximum_location])
```

```python
min_loc = min(city_dict, key=city_dict.get)
(min_loc, city_dict[min_loc])
```
